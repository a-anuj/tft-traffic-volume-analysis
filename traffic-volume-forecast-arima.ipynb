```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Metro_Interstate_Traffic_Volume.csv")
df['date_time'] = pd.to_datetime(df['date_time'])
df = df.sort_values('date_time')
df
```

```python
df['hour'] = df['date_time'].dt.hour
df['weekday'] = df['date_time'].dt.weekday
df['weekend'] = df['weekday'].isin([5,6]).astype(int)
df['month'] = df['date_time'].dt.month

df['rush_hour'] = df['hour'].isin([7,8,9,16,17,18]).astype(int)
```

```python
df
```

```python
df['weather_main'] = df['weather_main'].astype('category')
df['weather_description'] = df['weather_description'].astype('category')

# map to integer IDs for embedding layers
df['weather_main_id'] = df['weather_main'].cat.codes
df['weather_desc_id'] = df['weather_description'].cat.codes
```

```python
df
```

```python
df['holiday'] = df['holiday'].fillna("None")
df['holiday'] = df['holiday'].astype('category')
df['holiday_id'] = df['holiday'].cat.codes
```

```python
# Target column: traffic_volume
df['lag_1'] = df['traffic_volume'].shift(1)       # previous hour
df['lag_24'] = df['traffic_volume'].shift(24)     # previous day

# Rolling means
df['roll_3']  = df['traffic_volume'].rolling(3).mean()
df['roll_24'] = df['traffic_volume'].rolling(24).mean()
```

```python
df
```

```python
df = df.dropna()
df
```

```python
# ARIMA 
```

```python
target = df["traffic_volume"]
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
def stationarity_check(ts):
    result = adfuller(ts)
    print(f"ADF Statistic : {result[0]}")
    print(f"p-value:{result[1]}")
    print(f"Critical values : {result[4]}")
```

```python
stationarity_check(target)
```

### Need not apply any differencing since the target is already stationary

```python
best_aic = float('inf')
best_bic = float('inf')
best_order = None

for p in range(4):
    for d in range(2):
        for q in range(4):
            arima_model = ARIMA(target,order=(p,d,q))
            arima_results = arima_model.fit()

            current_aic = arima_results.aic
            current_bic = arima_results.bic

            if current_aic<best_aic and current_bic<best_bic:
                best_aic = current_aic
                best_bic = current_bic
                best_order = (p,d,q)
print(f"Best AIC : {best_aic} , Best BIC : {best_bic} , Best order : {best_order}")
```

```python
arima_model = ARIMA(target,order=best_order)
arima_results = arima_model.fit()
```

```python
residuals = arima_results.resid
stationarity_check(residuals)
```

```python
lb_test_stat = acorr_ljungbox(residuals, lags=40)
print(f'{lb_test_stat}')
```

### Clearly we can see that p-value <<< 0.05 which results in autocorrelation in the residuals. So we can conclude that this model(ARIMA) is not good for this multivariate forecasting.
